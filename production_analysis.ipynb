```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

```

**OPENING THE EXCEL FILE**

## PRODUCTION BY CONTINENTS

```python
df = pd.read_excel(
    r"C:\Users\dell\Downloads\6.1. Total_World_Production.xlsx",
    sheet_name="6.1.1.by Continents",
    header = 1
)

df.head()
```

**UNDERSTANDING THE DATA**

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna()

df.columns

df.rename(columns={})
```

```python
df_long = pd.melt(
    df,
    id_vars=["Continents"],
    var_name="Year",
    value_name="Production"
)
df_long 
```

```python
# convert columns to numeric types
df_long["Year"] = pd.to_numeric(df_long["Year"], errors ='coerce')
df_long["Production"] = pd.to_numeric(df_long["Production"], errors ='coerce')

# clean out any non-numeric rows 
df_long = df_long.dropna(subset=["Year","Production"])

#plotting code 
plt.figure(figsize=(15,6))
for c in df_long["Continents"].unique():

    subset = df_long[
        df_long["Continents"] == c
    ]

    plt.plot(
        subset["Year"],
        subset["Production"],
        label=c
    )
plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1))
plt.title("Production trend of countries")
plt.show()
```

**TOP PRODUCING CONTINENTS**

```python
# top producing continents
continent_total = (
    df_long
    .groupby("Continents")
    ["Production"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
continent_total
```

```python
# growth rate 
continent_growth = (
    df_long
    .groupby("Continents")
    ["Production"]
    .agg(["first","last"])
    .assign(growth_rate = lambda x: (x['last']-x['first'])/x['first']*100)
)
continent_growth
```

```python
#heatmap 
pivot = df_long.pivot(
    index="Continents",
    columns="Year",
    values="Production"
)
sns.heatmap(pivot,linewidths=0.3)
```

## POLITICAL STABILITY OF PRODUCER COUNTRIES ACCORDING TO WORLDWIDE GOVERENCE INDICATORS 

```python
df = pd.read_excel(
    r"C:\Users\dell\Downloads\6.1. Total_World_Production.xlsx",
    sheet_name="6.1.5.by Political Stability",
    header = 1
)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df_long = pd.melt(
    df,
    id_vars=["Political Stability of Producer Countries"],
    var_name="Year",
    value_name="Production"
)
df_long 
```

```python
# convert columns to numeric types
df_long["Year"] = pd.to_numeric(df_long["Year"], errors ='coerce')
df_long["Production"] = pd.to_numeric(df_long["Production"], errors ='coerce')

# clean out any non-numeric rows 
df_long = df_long.dropna(subset=["Year","Production"])

#plotting code 
plt.figure(figsize = (15,6))
for c in df_long["Political Stability of Producer Countries"].unique():

    subset = df_long[
        df_long["Political Stability of Producer Countries"] == c
    ]

    plt.plot(
        subset["Year"],
        subset["Production"],
        label=c
    )
plt.legend(
    loc='upper left',
    bbox_to_anchor=(1,1))
plt.title("Political Stability trend")
plt.show()
```

**In which years did STABLE COUNTRIES account for the largest share of global mineral production?**

```python
#stable years
stable = df[df.iloc[:,0]=="Stable"].iloc[0,1:]
total = df[df.iloc[:,0]=="Total"].iloc[0,1:]

stability_share = (stable/total)*100

top_years = stability_share.sort_values(
    ascending=False
).head(10)
print("TOP 10 STABLE YEARS")
print(top_years)
```

**In which years did UNSTABLE COUNTRIES account for the largest share of global mineral production?**

```python
#unstable years
unstable = df[df.iloc[:,0]=="Unstable"].iloc[0,1:]
total = df[df.iloc[:,0]=="Total"].iloc[0,1:]

stability_share = (unstable/total)*100

top_years = stability_share.sort_values(
    ascending=False
).head(10)
print("TOP 10 UNSTABLE YEARS")
print(top_years)
```

## TOTAL MINERAL PRODUCTION IN 2024 BY COUNTRIES 

```python
df = pd.read_excel(
    r"C:\Users\dell\Downloads\6.1. Total_World_Production.xlsx",
    sheet_name="6.1.8.1.Total Min. Prod. (t)",
    header = 1
)

df.head(170)
```

```python
## Which mineral does each country produce the most?
minerals = [ 'Iron and Ferro-Alloy Metals', 'Non-Ferrous Metals', 'Precious Metals (t)', 'Industrial Minerals (incl. Diam.)'
]

df['Largest_Mineral'] = df[minerals].idxmax(axis=1)

df[['Country','Largest_Mineral']].head()
```

```python
df['Largest_Mineral'].value_counts().plot(
    kind='bar',
    figsize=(10,5)
)

plt.title('Most Dominant Mineral Category by Country')
plt.show()
```

```python
minerals = [
    'Iron and Ferro-Alloy Metals',
    'Non-Ferrous Metals',
    'Precious Metals (t)',
    'Industrial Minerals (incl. Diam.)',
    'Mineral Fuels'
]

for mineral in minerals:

    country = df.loc[
        df[mineral].idxmax(),
        'Country'
    ]

    value = df[mineral].max()

    print(
        f"{mineral}: {country} ({value:,.0f})"
    )
```

```python
fuel_top = df.nlargest(15,'Mineral Fuels')

plt.figure(figsize=(12,6))
plt.barh(
    fuel_top['Country'],
    fuel_top['Mineral Fuels']
)

plt.title('Top Mineral Fuel Producing Countries')
plt.gca().invert_yaxis()
plt.show()
```

```python
top10 = df.nlargest(10,'Total (incl. Bauxite)')

minerals = [
    'Iron and Ferro-Alloy Metals',
    'Non-Ferrous Metals',
    'Precious Metals (t)',
    'Industrial Minerals (incl. Diam.)',
    'Mineral Fuels'
]

top10.set_index('Country')[minerals].plot(
    kind='bar',
    stacked=True,
    figsize=(14,6)
)

plt.title('Mineral Composition of Top Countries')
plt.show()
```

**Asia Will Continue to Dominate Production
Asia leads with 293.5 Billion mt (1984–2024), driven almost entirely by China**
China's dominance in mineral production is expected to continue through 2030, particularly in Industrial Minerals and Mineral Fuels. Policymakers and investors should monitor China's regulatory changes as they will directly impact global supply chains.

**Shift Toward Critical & Green Minerals
Data shows Mineral Fuels at 495 Billion mt as the top commodity. However, global energy transition trends suggest.**
While Mineral Fuels currently dominate production, demand for Non-Ferrous Metals (used in EV batteries, solar panels, and wind turbines) is projected to surge. Countries rich in Lithium, Cobalt, and Copper — particularly in Africa and South America — may see significant production growth post-2025.

**Africa & South America Are Emerging Producers
These continents are currently underrepresented in your data but hold vast untapped reserves.**
Africa and South America are likely to increase their share of global mineral production in the coming decade, driven by foreign investment and rising demand for battery metals. Their current low share represents an opportunity, not a ceiling.

**Political Stability is a Key Risk Factor**
As geopolitical tensions rise globally, production from unstable regions poses supply chain risks. Governments and corporations should diversify sourcing strategies and invest in minerals from politically stable nations to ensure supply security through 2030.

**Eastern Asia's Regional Dominance**
Eastern Asia's regional production advantage, largely driven by China, South Korea, and Japan, is unlikely to erode short-term. However, North America and Central Asia are positioned as the next growth hubs, particularly for rare earth elements and energy minerals.

**Overall Forward-Looking Statement**
Global mineral production is entering a transformative decade. The historical dominance of fossil fuel-based minerals will coexist with — and eventually yield to — demand for critical minerals supporting clean energy infrastructure. Nations that invest in sustainable extraction, political stability, and technological processing will emerge as the new leaders of the global mineral economy by 2035.
